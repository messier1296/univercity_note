## **課題3**
クラス編成問題において、数理最適化問題として表現可能な要件を１つ考えてください。

要件：各クラスに、学力試験の点数が370点以上の生徒を4人以上割り当てる。

この要件は、生徒ごとの点数とクラス割当変数を使って、各クラスに含まれる対象生徒数の下限制約として表現できる。

## **課題4**
課題3で考えた要件を制約式(目的関数)に表現し、次のセルに数式で書いてください（Tex記法推奨）。また、必要に応じて集合や定数を追加してください。

集合・定数・変数を次のように定義する。

- $I$：生徒集合
- $J$：クラス集合
- $H = \{ i \in I \mid score_i \ge 370 \}$：学力試験の点数が370点以上の生徒集合
- $x_{i,j}$：生徒 $i$ をクラス $j$ に割り当てるなら1、そうでなければ0をとる0-1変数

制約式は次のように表せる。

$$
\sum_{i \in H} x_{i,j} \ge 4 \quad (j \in J)
$$

## **課題5**
課題4で表現した制約式(目的関数)を追加した数理最適化モデルを以下のセルに実装し、実行してください。また、必要に応じてデータも作成してください。
- https://github.com/ohmsha/PyOptBook/tree/main/3.school にアップロードされているコードとデータを利用してかまいません。

```python
from pathlib import Path

import pandas as pd
import pulp

# ノートブックを授業ディレクトリから実行しても、リポジトリ直下から実行しても動くようにする。
data_dir = Path("data")
if not data_dir.exists():
    data_dir = Path("2026-spring/math_engineer/data")

students_df = pd.read_csv(data_dir / "students.csv")
pairs_df = pd.read_csv(data_dir / "student_pairs.csv")

classes = range(1, 9)
students = students_df["student_id"].tolist()
score = dict(zip(students_df["student_id"], students_df["score"]))
average_score = students_df["score"].mean()

male_students = students_df.loc[students_df["gender"] == 1, "student_id"].tolist()
female_students = students_df.loc[students_df["gender"] == 0, "student_id"].tolist()
leader_students = students_df.loc[students_df["leader_flag"] == 1, "student_id"].tolist()
support_students = students_df.loc[students_df["support_flag"] == 1, "student_id"].tolist()
high_score_students = students_df.loc[students_df["score"] >= 370, "student_id"].tolist()

prob = pulp.LpProblem("ClassAssignment", pulp.LpMinimize)
x = pulp.LpVariable.dicts("x", (students, classes), cat="Binary")

prob += 0

# 要件(1): 全生徒をちょうど1つのクラスに割り当てる。
for s in students:
    prob += pulp.lpSum(x[s][c] for c in classes) == 1

for c in classes:
    class_count = pulp.lpSum(x[s][c] for s in students)
    class_score = pulp.lpSum(score[s] * x[s][c] for s in students)

    # 要件(2): 各クラス39人以上40人以下。
    prob += class_count >= 39
    prob += class_count <= 40

    # 要件(3): 各クラスの男女それぞれ20人以下。
    prob += pulp.lpSum(x[s][c] for s in male_students) <= 20
    prob += pulp.lpSum(x[s][c] for s in female_students) <= 20

    # 要件(4): 各クラスの平均点を学年平均点±10点以内にする。
    prob += class_score <= (average_score + 10) * class_count
    prob += class_score >= (average_score - 10) * class_count

    # 要件(5): 各クラスにリーダー気質の生徒を2人以上。
    prob += pulp.lpSum(x[s][c] for s in leader_students) >= 2

    # 要件(6): 特別な支援が必要な生徒は各クラス1人以下。
    prob += pulp.lpSum(x[s][c] for s in support_students) <= 1

    # 課題4で追加した要件: 370点以上の生徒を各クラス4人以上。
    prob += pulp.lpSum(x[s][c] for s in high_score_students) >= 4

# 要件(7): 特定ペアは同一クラスにしない。
for row in pairs_df.itertuples(index=False):
    for c in classes:
        prob += x[row.student_id1][c] + x[row.student_id2][c] <= 1

status = prob.solve(pulp.PULP_CBC_CMD(msg=False))
print("Status:", pulp.LpStatus[status])
print("学年平均点:", round(average_score, 2))

assignments = []
for c in classes:
    assigned = [s for s in students if pulp.value(x[s][c]) > 0.5]
    assignments.append(
        {
            "class": c,
            "students": len(assigned),
            "male": sum(s in male_students for s in assigned),
            "female": sum(s in female_students for s in assigned),
            "average_score": round(sum(score[s] for s in assigned) / len(assigned), 2),
            "leaders": sum(s in leader_students for s in assigned),
            "supports": sum(s in support_students for s in assigned),
            "score_370_or_more": sum(s in high_score_students for s in assigned),
        }
    )

summary_df = pd.DataFrame(assignments)
display(summary_df)
```
