```julia
f(x) = -x^0.75 + x -5
EPSILON = 1e-10

function bisection(f,a,b,EPSILON)
    if a >= b 
        return
    end
    
    iteration = 0
    while abs(b-a) > EPSILON && abs(f(b)) > EPSILON
        iteration += 1; 
        c = (a + b) / 2;
        if f(a) * f(c) < 0
            b = c;
        else
            a = c;
        end
    end

    return iteration
end

a = 10;
b = 20;


print(bisection(f,a,b,EPSILON));
```

```julia
function derive(f,x)
    return (f(x + EPSILON) - f(x)) / EPSILON
    
end

function newton(f,x,EPSILON)
    iteration = 0;
    while abs(f(x)) >= EPSILON
        iteration += 1;
        x2 = x - f(x) / derive(f,x);
        if abs(x - x2) >= EPSILON
            x = x2;
        else 
            break
        end        
    end
    return iteration
end

print(newton(f,10,EPSILON))
```

```julia
using Plots

eps = 10.0 .^ range(-10,-1;length=1000);

iteration_b = [
    bisection(f,10.0,20.0,e)
    for e in  eps
]
iteration_n = [
    newton(f,10,e)
    for e in eps
]

plot(
    eps,
    iteration_b,
    label = "bisection",
    xlabel="eps",
    ylabel="iteration",
    xscale=:log10,
)

plot!(
    eps,
    iteration_n,
    label="newton",
)
```
